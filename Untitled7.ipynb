````python
# ÉTAPE 2 — NETTOYAGE DES COLONNES NUMÉRIQUES
# (Budget et Prix contiennent des valeurs comme "5000f", "10.000f", "RAS")
# ============================================================

def nettoyer_numerique(valeur):
    """
    Nettoie une valeur numérique mal formatée :
    - "5000f"   → 5000.0
    - "10.000f" → 10000.0
    - "RAS"     → NaN (sera imputé après)
    - NaN       → NaN
    """
    if pd.isna(valeur):
        return np.nan
    valeur_str = str(valeur).strip()
    # Supprimer les lettres (f, F, FCFA...)
    valeur_str = ''.join(c for c in valeur_str if c.isdigit() or c in '.,')
    # Remplacer le point séparateur de milliers : "10.000" → "10000"
    if '.' in valeur_str and len(valeur_str.split('.')[-1]) == 3:
        valeur_str = valeur_str.replace('.', '')
    # Remplacer la virgule décimale
    valeur_str = valeur_str.replace(',', '.')
    try:
        return float(valeur_str)
    except (ValueError, TypeError):
        return np.nan  # "RAS" et autres textes → NaN
````

````python
# Appliquer le nettoyage
col_budget = 'Quel est votre budget approximatif pour les emballages alimentaires par mois ?'
col_prix   = 'A combien vous reviens la centaine des plats que vous utiliser actuellement?'

df[col_budget] = df[col_budget].apply(nettoyer_numerique)
df[col_prix]   = df[col_prix].apply(nettoyer_numerique)

print("\n=== APRÈS NETTOYAGE : Budget et Prix ===")
print(df[[col_budget, col_prix]].describe())
````

## Préférences pour le Nombre de Compartiments




````python
compartment_options = ['1 seul compartiment', '2 compartiments', '3 compartiments', '4 compartiments']

# Initialize a dictionary to store counts
compartment_counts = {option: 0 for option in compartment_options}

# Iterate over the column and count preferences
for index, row in df.iterrows():
    preferences_str = row[' Vous voulez des  emballages alimentaires avec combien de compartilment']
    if pd.notna(preferences_str):
        individual_preferences = [pref.strip() for pref in preferences_str.split(',')]
        for pref in individual_preferences:
            # Handle cases where user might just type '1 compartiment' instead of '1 seul compartiment'
            if '1 compartiment' in pref and '1 seul compartiment' in compartment_options:
                compartment_counts['1 seul compartiment'] += 1
            elif '2 compartiments' in pref and '2 compartiments' in compartment_options:
                compartment_counts['2 compartiments'] += 1
            elif '3 compartiments' in pref and '3 compartiments' in compartment_options:
                compartment_counts['3 compartiments'] += 1
            elif '4 compartiments' in pref and '4 compartiments' in compartment_options:
                compartment_counts['4 compartiments'] += 1

# Create a DataFrame from the counts
compartment_df = pd.DataFrame(compartment_counts.items(), columns=['Nombre de compartiments', 'Nombre de préférences'])

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=compartment_df, x='Nombre de compartiments', y='Nombre de préférences', hue='Nombre de compartiments', palette='viridis', legend=False)
plt.title('Préférences pour le nombre de compartiments')
plt.xlabel('Nombre de compartiments')
plt.ylabel('Nombre de préférences')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

## Catégoriser le prix actuel

### Subtask:
Créer des catégories de prix (Faible, Moyen, Élevé) à partir de la colonne 'A combien vous reviens la centaine des plats que vous utiliser actuellement?' pour faciliter l'analyse.


````python
col_prix = 'A combien vous reviens la centaine des plats que vous utiliser actuellement?'

# Calculate the 33rd and 66th percentiles
pct_33 = df[col_prix].quantile(0.33)
pct_66 = df[col_prix].quantile(0.66)

print(f"33rd percentile (Faible/Moyen threshold): {pct_33:,.0f} FCFA")
print(f"66th percentile (Moyen/Élevé threshold) : {pct_66:,.0f} FCFA")

# Create the 'Catégorie de Prix Actuel' column
def categorize_price(price):
    if price <= pct_33:
        return 'Faible'
    elif price <= pct_66:
        return 'Moyen'
    else:
        return 'Élevé'

df['Catégorie de Prix Actuel'] = df[col_prix].apply(categorize_price)

# Display the distribution of the new column
print("\nDistribution de la nouvelle colonne 'Catégorie de Prix Actuel':")
print(df['Catégorie de Prix Actuel'].value_counts())
````

## Explorer les Distributions des Variables Clés

### Subtask:
Analyze the distribution of objectives customers seek with food packaging from the column 'Quels sont les principaux objectifs que vous recherchez avec les emballages alimentaires utilisés?'.


**Reasoning**:
First, I need to define a function that can take a string from the 'Quels sont les principaux objectifs que vous recherchez avec les emballages alimentaires utilisés?' column, split it by commas, clean up any leading/trailing whitespace, and return a list of individual objectives. This function will be crucial for processing the raw data into a usable format for analysis.



````python
def split_and_clean_objectives(objectives_str):
    """
    Splits a comma-separated string of objectives, cleans each objective,
    and returns a list of cleaned objectives.
    """
    if pd.isna(objectives_str):
        return []
    return [obj.strip() for obj in str(objectives_str).split(',') if obj.strip()]

print("Function `split_and_clean_objectives` defined.")
````

**Reasoning**:
Now that the helper function is defined, I will apply it to the 'objectives' column to extract all individual objectives and then count their frequencies to understand the distribution of customer goals.



````python
import itertools

# Identify the column containing objectives
objectives_col = 'Quels sont les principaux objectifs que vous recherchez avec les  emballages alimentaires  utilisés?'

# Apply the function to the column and flatten the list of lists
all_objectives = list(itertools.chain(*df[objectives_col].apply(split_and_clean_objectives)))

# Count the occurrences of each objective
objectives_counts = pd.Series(all_objectives).value_counts()

print("Distribution des objectifs recherchés par les clients :\n")
print(objectives_counts.head(10))
````
